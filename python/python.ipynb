### Built-In Functions

---

<div class="alert alert-block alert-warning">
So, let's start off with  built-in functions in Python. Hopefully, everyone's heard of a function, but that's okay if you haven't.

<br />

A function is just a name assigned to multiple lines of Python code. Where, each line of Python code is responsible for doing its own individual task. So, just as an example, we may have one line of Python that prints out some text to a string, we may have another line that adds up some numbers together. There are an endless number of things we can do in a single line of code, but just know each line does one unique thing.

<br />

Anyways, just to summarize, a function is a name assigned to lines of Python code. And, after creating a function, we can call that name to run those lines of code anywhere in our Python file. So, basically a function just lets us reuse lines of Python code across a single Python file.

```python
print("hello everyone!")
```

### Comparing Functions and Variables

---

- Variables
    - A name assigned to a **single line of code**
    - Can reuse variables across a single Python file
- Functions
    - A name assigned to **multiple lines of code**
    - Can reuse functions across a single Python file

<div class="alert alert-block alert-warning">
Now, let me take a moment right now to review some of the information I've touched on. Specifically, let me outline some of the differences between variables and functions, because I know I've only briefly talked about them.

<br />

Remember, a <i>variable</i> is a name assigned to a single line of Python code, whereas a <i>function</i> is a name assigned to multiple lines of Python code.

<br />

Also, a <i>variable</i> lets us reuse a single line of Python code across a single Python file, whereas a <i>function</i> lets us reuse multiple lines of Python code across a single Python file.

<br />

Keep in mind, after we create a function, we can call its name to run its <i>lines of code</i> anywhere in our Python file. Similarly, after we create a variable, we can call its name to run that <i>single line of code</i> anywhere in our Python file.
</div>

### Custom Functions

---

<div class="alert alert-block alert-warning">
So, now that we've gone over the differences between functions and variables, and we've introduced this idea of built-in functions, let's go deeper into this idea of functions. Specifically, we'll introduce this idea of custom functions in Python.

<br />

And really, a custom function is fairly self-explanatory. It's just a function we create ourselves, and wasn't handed to us by the developers of Python when they first programmed the language.
</div>

```python
def print_avg(a, b, c):
    sum_of_nums = a + b + c
    avg = sum_of_nums / 3
    return avg
```

<div class="alert alert-block alert-warning">
Just as an example, here, we've created a custom function and named it <i>print_avg</i>. Notice, the <i>def</i> keyword is reserved for creating a new function. Also, we've declared two variables called <i>sum_of_nums</i> and <i>avg</i>. For sum_of_nums, we add up a, b, and c, then save it. For avg, we take the sum, divide it by 3, then save it. This shouldn't be new to us, since we did this earlier with our built-in print function.
</div>

```python
print_avg(1, 2, 3)
```

<div class="alert alert-block alert-warning">
Now, we can use our custom function! Pretty cool.
</div>

```python
print((1+2+3) / 3)
```

<div class="alert alert-block alert-warning">
And just notice, we can get a similar output using the built-in print function. It's just, now our custom function is much more readable and customizable.
</div>

```python
print_avg(1, '2', 3)
```

<div class="alert alert-block alert-warning">
And I just want to bring up a point about data types again. When we create a custom function, our inputs must be Int data types. So, we can't input String data types. It's all about making sure our data types are the same data types as the ones we specified when we declared our input values in the custom function.
</div>

# What is a Package?

<div class="alert alert-block alert-warning">
Now that we've seen what some Python code looks like, and we know that we can write this code in a Python file, we're ready to start looking at something called <i>packages</i>. 

<br />
    
So, a package is just a folder of python files, or a collection of Python files. In Python, we can use other people's python files within a package. To do this, we must <i>import</i> these files into our own python script, using the <i>import</i> keyword. More specifically, if we import a package, we're basically reading in all of the functions from that package.

<br />

Alright, so we should probably be asking ourselves right now, how does Python know where to import a package from. Where exactly does a package live? Well, Python basically has its own site, and they allow anyone to save their package on their site for free. By doing this, we can create our own Python files, then save them on Python's site as a package. From that point, anyone can download our package from Python's site by calling <i>import</i>. The best part is anyone can import and use our package in seconds for free. 
</div>

<div class="alert alert-block alert-warning">
Just to help make sure we all have at least a high level understanding of what a package is, I'll motivate a Python package using an analogy, comparing a Python package to a multi-panel painting being sold on Etsy.

Let's say we just finished creating our own handmade multi-panel painting. Now, we really love the painting, and so we think many other people would like to have it in their home too. Then, let's say we send our painting to Etsy and have them distribute it for us. So, we package up our panels, send our package to Etsy, and they handle all of the behind-the-scenes logistics. Now, whenever someone wants to have our painting, they can just buy replicas of them from Etsy!

<br />

In this analogy, Etsy here acts as the official Python site distributing our package. Then, the Python package is the multi-panel painting, and each of the Python files in that package are the individual panels.

<br />

Now, I should note here, each panel is a separate thing. But, the key is theyre sold together. It wouldnt make any sense for Etsy to send individual panels to anyone. Instead, Etsy will only send people the entire multi-panel painting. This is similar to Python exports packages to people when they import them.

There are some reasons why this analogy isn't perfect. Like, in a normal transaction with Etsy, the customer would need to wait days to receive their painting. ### Python vs. SQL

---

- Python and SQL both used for:
    - Transforming data
    - Loading in various types of data
- Python can do these things better:
    - Doing more complex analysis
    - Processing smaller, raw data
- SQL can do these things better:
    - Transforming data already loaded into the database
    - Transforming data that can't fit in memory

<div class="alert alert-block alert-warning">
So, when you first hear that Python has some of these packages, you may reasonably ask yourself <i>why would I ever want to learn Python when I can do some of this in SQL</i>? Yeah, that's a very fair question, and the answer to that question just depends on the person and their needs.

<br />

SQL and pandas can fill similar high-level roles. Specifically, they can both be used for transforming data, and they can load lots of different types of data. But they do have their obvious differences.

<br />

Specifically, pandas can be better for doing more complex analysis, because pandas supports functions used for more complex analysis, whereas a lot of SQL languages don't really support similar functions, like gathering, pivoting, summary functions. Also, pandas can be used for processing smaller and more raw data. This is when we load data from a SQL database, but it's not quite summarized or aggregated the way we want it.

<br />

But, I'm not trying to say there's one best solution. They each have their own roles. Specifically, SQL can be better for transforming data already loaded in the database. Also, SQL can be better for transforming larger data that can't be loaded into memory. Just so you know, pandas loads all of the data in memory on your own computer, and this memory is very limited. Most SQL databases will load data in memory too, but they just have much more memory or storage available for processing data already stored in databases. So, SQL can be much more effecient if our data is already in our database, and we don't need much more complicated analytical functions than the supported SQL functions.
</div>

### Python vs. Tableau

---

- Python and Tableau both used for:
    - Creating interactive visualizations
    - Creating prototypes
    - Exploring dataset
- Python can do these things better:
    - Building visualizations for free
    - Building visualizations with more flexibility
- Tableau can do these things better:
    - Building visualizations where clean data already connected
    - Exploring datasets

<div class="alert alert-block alert-warning">
Similar to our question about using SQL vs. Python, we may ask ourselves <i>why we can't just use Tableau here</i>? And you'll get a similar answer: you can! Again, Python isn't a 1-for-1 replacement of SQL or Tableau. It can be used as a subsitute for certain SQL and Tableau use-cases

<br />

So, just like SQL, there is overlap with plotly and Tableau. Specifically, they both can create interactive visualizations, they both can be used for prototyping, and they both can be used for exploring datasets.

Then, plotly can be better for building visualizations for free, since plotly is completely free, whereas you need a license to use Tableau. Lastly, Tableau is great for cleaning data we're already connected to.
